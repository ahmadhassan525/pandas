```python
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
a = [1,2,3,4,5,6,7]
b = [1,2,4,5,6,9,20]
c = [20,9,8,6,7,4,2]
```

```python
plt.plot(a,b,label='this is b')
plt.plot(a,c,label='this is c')
plt.xlabel("a")
plt.ylabel('b and c')
plt.legend()
plt.show()
```

```python
df = pd.read_csv('../data/countries.csv')
```

```python
df.sample(10)
```

```python
us  = df[df.country == 'United States']
china = df[df.country == 'China']
us_pop = (us.population / 10**6)
china_pop = (china.population / 10**6)
```

```python
plt.plot(us.year, us_pop , label='United States')
plt.plot(china.year,china_pop,label = "China")
plt.xlabel('Years')
plt.ylabel('Population')
plt.title('population count / 10**6')
plt.legend()
plt.show()
```

```python
us_pop_max_div = (us.population / us.population.iloc[0] *100)
china_pop_max_div = (china.population / china.population.iloc[0] *100)
```

```python
plt.plot(us.year,us_pop_max_div,label="us population")
plt.plot(china,year,china_pop_max_div,label="china population")
plt.title('population / population first * 100')
plt.xlabel('Year')
plt.ylabel('Popualtion')
plt.legend()
plt.show()
```
